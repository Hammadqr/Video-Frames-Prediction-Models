```python

# Import necessary libraries
import os
import cv2
import numpy as np
import kagglehub
import matplotlib.pyplot as plt
from tqdm import tqdm
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, ConvLSTM2D, BatchNormalization, Conv3D, Lambda
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping
from tensorflow.keras.utils import Sequence

# (Optional) Enable mixed precision if supported
from tensorflow.keras import mixed_precision
# mixed_precision.set_global_policy('mixed_float16')  # Uncomment if you want to enable mixed precision

# Define selected classes
selected_classes = ["Biking", "GolfSwing", "PushUps", "HandstandWalking", "JumpingJack"]

# Download the dataset using kagglehub
path = kagglehub.dataset_download("matthewjansen/ucf101-action-recognition")
print("Path to dataset files:", path)

# Define paths for train, validation, and test sets
train_path = "/kaggle/input/ucf101-action-recognition/train"
val_path = "/kaggle/input/ucf101-action-recognition/val"
test_path = "/kaggle/input/ucf101-action-recognition/test"

# Define output paths for processed data
output_path_train = "/kaggle/working/processed_ucf101_train"
output_path_val = "/kaggle/working/processed_ucf101_val"
output_path_test = "/kaggle/working/processed_ucf101_test"

# Preprocessing parameters
target_size = (64, 64)  # Width, Height
grayscale = True        # Convert frames to grayscale

def create_output_dirs(paths):
    """
    Create output directories if they do not exist.
    """
    for path in paths:
        if not os.path.exists(path):
            os.makedirs(path)
            print(f"Created directory: {path}")
        else:
            print(f"Directory already exists: {path}")

# Create output directories for train, val, and test
create_output_dirs([output_path_train, output_path_val, output_path_test])

def process_video(video_path, target_size=(64, 64), grayscale=True):
    """
    Process a video to extract frames, resize them, and convert to grayscale if needed.

    Args:
        video_path (str): Path to the video file.
        target_size (tuple): Desired frame size (width, height).
        grayscale (bool): Whether to convert frames to grayscale.

    Returns:
        np.ndarray: Array of processed frames.
    """
    cap = cv2.VideoCapture(video_path)
    frames = []
    
    while True:
        ret, frame = cap.read()
        if not ret:
            break
        # Resize frame
        frame = cv2.resize(frame, target_size)
        if grayscale:
            frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
            frame = frame[..., np.newaxis]  # Add channel dimension
        frames.append(frame)
    
    cap.release()
    
    if len(frames) == 0:
        return None  # Return None if no frames were extracted
    
    frames = np.array(frames)
    
    return frames

def preprocess_and_save(split_path, output_path, selected_classes, target_size=(64, 64), grayscale=True):
    """
    Preprocess videos and save them as .npy files.

    Args:
        split_path (str): Path to the dataset split (train/val/test).
        output_path (str): Path to save the processed .npy files.
        selected_classes (list): List of class names to process.
        target_size (tuple): Desired frame size (width, height).
        grayscale (bool): Whether to convert frames to grayscale.
    """
    for class_name in selected_classes:
        class_input_path = os.path.join(split_path, class_name)
        class_output_path = os.path.join(output_path, class_name)
        
        # Create class-specific output directory
        if not os.path.exists(class_output_path):
            os.makedirs(class_output_path)
            print(f"Created directory: {class_output_path}")
        else:
            print(f"Directory already exists: {class_output_path}")
        
        # Iterate over all .avi files in the class directory
        video_files = [f for f in os.listdir(class_input_path) if f.endswith('.avi')]
        
        print(f"Processing class '{class_name}' with {len(video_files)} videos.")
        
        for video_file in tqdm(video_files, desc=f"Processing {class_name}"):
            video_path = os.path.join(class_input_path, video_file)
            frames = process_video(video_path, target_size, grayscale)
            
            if frames is not None:
                # Save the frames as a .npy file with the same base name as the video
                video_name = os.path.splitext(video_file)[0]
                output_file = os.path.join(class_output_path, f"{video_name}.npy")
                np.save(output_file, frames)
            else:
                print(f"Warning: No frames extracted from {video_path}")

# Preprocess and save training data
print("Starting preprocessing for Training Data...")
preprocess_and_save(train_path, output_path_train, selected_classes, target_size, grayscale)
print("Training data preprocessing completed.\n")

# Preprocess and save validation data
print("Starting preprocessing for Validation Data...")
preprocess_and_save(val_path, output_path_val, selected_classes, target_size, grayscale)
print("Validation data preprocessing completed.\n")

# Preprocess and save testing data
print("Starting preprocessing for Testing Data...")
preprocess_and_save(test_path, output_path_test, selected_classes, target_size, grayscale)
print("Testing data preprocessing completed.\n")

def visualize_sample(output_path, class_name, video_index=0, grayscale=True):
    """
    Visualize a sample video from the preprocessed data.

    Args:
        output_path (str): Path to the processed data.
        class_name (str): Name of the class to visualize.
        video_index (int): Index of the video within the class.
        grayscale (bool): Whether the frames are in grayscale.
    """
    class_path = os.path.join(output_path, class_name)
    video_files = [f for f in os.listdir(class_path) if f.endswith('.npy')]
    
    if len(video_files) == 0:
        print(f"No .npy files found in {class_path}")
        return
    
    if video_index >= len(video_files):
        print(f"Video index {video_index} out of range for class '{class_name}'.")
        return
    
    selected_video = video_files[video_index]
    video_path = os.path.join(class_path, selected_video)
    frames = np.load(video_path)
    
    num_frames = frames.shape[0]
    plt.figure(figsize=(15, 3))
    
    # Display first 5 frames or all frames if less than 5
    num_display = min(5, num_frames)
    for i in range(num_display):
        plt.subplot(1, num_display, i+1)
        if grayscale:
            plt.imshow(frames[i, :, :, 0], cmap='gray')
        else:
            plt.imshow(frames[i].astype(np.uint8))
        plt.axis('off')
        plt.title(f"Frame {i+1}")
    
    plt.suptitle(f"Sample Video: {selected_video} ({class_name})")
    plt.show()

# Example: Visualize the first video of the 'Biking' class in training data
visualize_sample(output_path_train, "Biking", video_index=0, grayscale=grayscale)

class DataGenerator(Sequence):
    """
    Custom data generator for loading video data from .npy files in batches.
    """
    def __init__(self, file_paths, batch_size=32, input_frames=10, pred_frames=5, 
                 target_size=(64, 64), channels=1, shuffle=True):
        self.file_paths = file_paths
        self.batch_size = batch_size
        self.input_frames = input_frames
        self.pred_frames = pred_frames
        self.target_size = target_size
        self.channels = channels
        self.shuffle = shuffle
        self.on_epoch_end()
    
    def __len__(self):
        """Denotes the number of batches per epoch"""
        return int(np.floor(len(self.file_paths) / self.batch_size))
    
    def __getitem__(self, index):
        """Generate one batch of data"""
        # Generate file paths for the current batch
        batch_files = self.file_paths[index*self.batch_size:(index+1)*self.batch_size]
        
        # Initialize lists
        X_batch = []
        Y_batch = []
        
        for file_path in batch_files:
            frames = np.load(file_path)
            total_frames = frames.shape[0]
            if total_frames < self.input_frames + self.pred_frames:
                continue  # Skip if not enough frames
            
            # Randomly select a start index for the window
            start_idx = np.random.randint(0, total_frames - self.input_frames - self.pred_frames + 1)
            input_seq = frames[start_idx:start_idx + self.input_frames]
            output_seq = frames[start_idx + self.input_frames:start_idx + self.input_frames + self.pred_frames]
            
            X_batch.append(input_seq)
            Y_batch.append(output_seq)
        
        # Convert lists to numpy arrays and normalize
        X_batch = np.array(X_batch).astype('float32') / 255.
        Y_batch = np.array(Y_batch).astype('float32') / 255.
        
        return X_batch, Y_batch
    
    def on_epoch_end(self):
        """Updates indexes after each epoch"""
        if self.shuffle:
            np.random.shuffle(self.file_paths)

def get_all_npy_files(output_path, selected_classes):
    """
    Retrieve all .npy file paths for the specified classes.
    
    Args:
        output_path (str): Base path where processed data is stored.
        selected_classes (list): List of class names.
    
    Returns:
        list: List of full file paths to .npy files.
    """
    all_files = []
    for class_name in selected_classes:
        class_path = os.path.join(output_path, class_name)
        video_files = [os.path.join(class_path, f) for f in os.listdir(class_path) if f.endswith('.npy')]
        all_files.extend(video_files)
    return all_files

# Retrieve file paths
train_files = get_all_npy_files(output_path_train, selected_classes)
val_files = get_all_npy_files(output_path_val, selected_classes)

print(f"Number of training files: {len(train_files)}")
print(f"Number of validation files: {len(val_files)}")

# Define batch size
batch_size = 8  # Adjust based on available GPU memory

# Create generators
training_generator = DataGenerator(
    file_paths=train_files,
    batch_size=batch_size,
    input_frames=10,
    pred_frames=5,
    target_size=target_size,
    channels=1,
    shuffle=True
)

validation_generator = DataGenerator(
    file_paths=val_files,
    batch_size=batch_size,
    input_frames=10,
    pred_frames=5,
    target_size=target_size,
    channels=1,
    shuffle=False
)

def build_predrnn_model(input_frames=10, height=64, width=64, channels=1, pred_frames=5):
    """
    Build an encoder-decoder PredRNN model using ConvLSTM2D layers.
    
    Args:
        input_frames (int): Number of input frames.
        height (int): Height of each frame.
        width (int): Width of each frame.
        channels (int): Number of channels in each frame (1 for grayscale).
        pred_frames (int): Number of frames to predict.
    
    Returns:
        keras.Model: Compiled PredRNN model.
    """
    # Encoder
    encoder_input = Input(shape=(input_frames, height, width, channels), name='encoder_input')
    
    x = ConvLSTM2D(filters=64, kernel_size=(3,3),
                  padding='same', return_sequences=True, activation='relu', name='encoder_conv_lstm_1')(encoder_input)
    x = BatchNormalization(name='encoder_batch_norm_1')(x)
    
    x = ConvLSTM2D(filters=64, kernel_size=(3,3),
                  padding='same', return_sequences=True, activation='relu', name='encoder_conv_lstm_2')(x)
    x = BatchNormalization(name='encoder_batch_norm_2')(x)
    
    x = ConvLSTM2D(filters=64, kernel_size=(3,3),
                  padding='same', return_sequences=False, activation='relu', name='encoder_conv_lstm_3')(x)
    x = BatchNormalization(name='encoder_batch_norm_3')(x)
    
    # Latent Space: (batch_size, height, width, channels)
    
    # Repeat the latent space for each predicted frame using a Lambda layer
    def repeat_time(x, rep=5):
        # Expand dims to add time dimension
        x = tf.expand_dims(x, 1)  # (batch_size, 1, height, width, channels)
        # Repeat along time dimension
        return tf.repeat(x, rep, axis=1)  # (batch_size, rep, height, width, channels)
    
    x = Lambda(lambda x: repeat_time(x, pred_frames), name='repeat_time')(x)  # (batch_size, pred_frames, height, width, channels)
    
    # Decoder
    x = ConvLSTM2D(filters=64, kernel_size=(3,3),
                  padding='same', return_sequences=True, activation='relu', name='decoder_conv_lstm_1')(x)
    x = BatchNormalization(name='decoder_batch_norm_1')(x)
    
    x = ConvLSTM2D(filters=64, kernel_size=(3,3),
                  padding='same', return_sequences=True, activation='relu', name='decoder_conv_lstm_2')(x)
    x = BatchNormalization(name='decoder_batch_norm_2')(x)
    
    x = ConvLSTM2D(filters=64, kernel_size=(3,3),
                  padding='same', return_sequences=True, activation='relu', name='decoder_conv_lstm_3')(x)
    x = BatchNormalization(name='decoder_batch_norm_3')(x)
    
    # Output layer: Conv3D to map to desired channels
    decoder_output = Conv3D(
        filters=channels,
        kernel_size=(3,3,3),
        activation='sigmoid',
        padding='same',
        data_format='channels_last',
        dtype='float32',  # Ensure output is float32
        name='decoder_output'
    )(x)
    
    # Define the model
    model = Model(inputs=encoder_input, outputs=decoder_output, name='PredRNN_Encoder_Decoder')
    
    # Compile the model
    optimizer = Adam(learning_rate=1e-4)
    model.compile(optimizer=optimizer, loss='mse', metrics=['mae'])
    
    model.summary()
    return model

# Build the model
predrnn_model = build_predrnn_model()

# Define callbacks
callbacks = [
    EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True),
    ModelCheckpoint('predrnn_model.keras', save_best_only=True)
]

# Verify model output shape
sample_input = np.random.rand(1, 10, 64, 64, 1).astype('float32') / 255.
sample_output = predrnn_model.predict(sample_input)
print("Sample output shape:", sample_output.shape)  # Should be (1, 5, 64, 64, 1)

# Train the model using generators
history = predrnn_model.fit(
    training_generator,
    epochs=10,                    # Adjust epochs as needed
    validation_data=validation_generator,
    callbacks=callbacks,
)

# Save the final model
predrnn_model.save("predrnn_model_final.keras")
```

```python
# New Cell: Plot Training History

# Import matplotlib if not already imported
import matplotlib.pyplot as plt

# Print available metrics in history.history
print("Available metrics in history.history:", history.history.keys())

def plot_training_history(history):
    """Plot the training and validation loss and MAE."""
    plt.figure(figsize=(12, 5))

    # Plot Training and Validation Loss (MSE)
    plt.subplot(1, 2, 1)
    plt.plot(history.history['loss'], label='Training Loss (MSE)')
    plt.plot(history.history['val_loss'], label='Validation Loss (MSE)')
    plt.title('Training and Validation Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss (MSE)')
    plt.legend()
    plt.grid(True)

    # Plot Training and Validation MAE
    plt.subplot(1, 2, 2)
    plt.plot(history.history['mae'], label='Training MAE')
    plt.plot(history.history['val_mae'], label='Validation MAE')
    plt.title('Training and Validation MAE')
    plt.xlabel('Epochs')
    plt.ylabel('Mean Absolute Error')
    plt.legend()
    plt.grid(True)

    plt.tight_layout()
    plt.show()

# Call the plotting function
plot_training_history(history)
```

```python
# New Cell: Compute and Plot SSIM on Validation Data

import tensorflow as tf
import matplotlib.pyplot as plt

def compute_ssim_on_validation(model, validation_generator):
    """
    Compute SSIM between model predictions and ground truth on validation data.
    Returns a list of SSIM scores.
    """
    ssim_scores = []
    
    # Iterate over all batches in the validation generator
    for i in range(len(validation_generator)):
        X_batch, Y_batch = validation_generator[i]
        
        # Predict the future frames
        Y_pred = model.predict(X_batch)
        
        # Ensure the predictions and ground truths are in the same range [0, 1]
        Y_pred = np.clip(Y_pred, 0., 1.)
        
        # Reshape to combine batch and sequence dimensions
        # Y_pred and Y_batch have shape (batch_size, pred_frames, height, width, channels)
        batch_size, pred_frames, height, width, channels = Y_pred.shape
        
        Y_pred_reshaped = Y_pred.reshape((batch_size * pred_frames, height, width, channels))
        Y_batch_reshaped = Y_batch.reshape((batch_size * pred_frames, height, width, channels))
        
        # Compute SSIM for each image in the batch
        ssim = tf.image.ssim(Y_batch_reshaped, Y_pred_reshaped, max_val=1.0)
        ssim_scores.extend(ssim.numpy())
        
    return ssim_scores

# Compute SSIM on validation data
ssim_scores = compute_ssim_on_validation(predrnn_model, validation_generator)

# Compute mean SSIM
mean_ssim = np.mean(ssim_scores)
print(f"Mean SSIM on validation data: {mean_ssim:.4f}")

# Plot histogram of SSIM scores
plt.figure(figsize=(8, 6))
plt.hist(ssim_scores, bins=50, density=True, alpha=0.7, color='blue')
plt.title('Histogram of SSIM Scores on Validation Data')
plt.xlabel('SSIM')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```

```python
import cv2
import numpy as np

def predicted_frames_to_video(input_seq, predicted_seq, output_path='predicted_video.mp4', fps=10):
    """
    Convert predicted frames to an MP4 video.
    
    Args:
        input_seq (np.ndarray): Input sequence of frames, shape (10, 64, 64, 1)
        predicted_seq (np.ndarray): Predicted sequence of frames, shape (5, 64, 64, 1)
        output_path (str): Path to save the output video
        fps (int): Frames per second for the output video
    """
    # Combine input and predicted sequences
    combined_frames = np.concatenate([input_seq, predicted_seq], axis=0)
    
    # Ensure frames are in uint8 format
    combined_frames = combined_frames.astype(np.uint8).squeeze()
    
    # Create VideoWriter object
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    out = cv2.VideoWriter(output_path, fourcc, fps, (64, 64), isColor=False)
    
    # Write frames to video
    for frame in combined_frames:
        out.write(frame)
    
    # Release the VideoWriter
    out.release()
    
    print(f"Video saved to {output_path}")
    print(f"Total frames in video: {combined_frames.shape[0]}")

# Usage example
predicted_frames_to_video(
    input_sequence_denorm, 
    predicted_frames_denorm, 
    output_path='/kaggle/working/jumping_jack_prediction.mp4'
)
```

```python
# Import Gradio and other necessary libraries
import gradio as gr
import os
import numpy as np
import matplotlib.pyplot as plt
import io
from PIL import Image

def get_available_videos(output_path, selected_classes):
    """
    Get a dictionary of available videos for each class.

    Args:
        output_path (str): Base path where processed data is stored.
        selected_classes (list): List of class names.

    Returns:
        dict: Dictionary with class names as keys and list of video filenames as values.
    """
    video_dict = {}
    for class_name in selected_classes:
        class_path = os.path.join(output_path, class_name)
        video_files = [f for f in os.listdir(class_path) if f.endswith('.npy')]
        video_dict[class_name] = video_files
    return video_dict

# Define the paths and classes
output_path = output_path_train  # Using training data; you can change to validation or test data
video_dict = get_available_videos(output_path, selected_classes)

def predict_future_frames_from_npy(class_name, video_file):
    """
    This function takes a class name and a video file name, loads the .npy file,
    processes it, runs the predrnn_model to predict future frames, and returns the predicted frames.
    """
    video_path = os.path.join(output_path, class_name, video_file)
    
    # Load the video frames
    frames = np.load(video_path)
    if frames.shape[0] < 10:
        return "The selected video must have at least 10 frames."
    
    # Take the first 10 frames as input
    input_sequence = frames[:10]  # Shape: (10, 64, 64, 1)
    input_sequence = input_sequence.astype('float32') / 255.0
    input_sequence = np.expand_dims(input_sequence, axis=0)  # Shape: (1, 10, 64, 64, 1)
    
    # Run the model to predict future frames
    predicted_frames = predrnn_model.predict(input_sequence)
    predicted_frames = predicted_frames[0] * 255.0  # Shape: (5, 64, 64, 1)
    
    # Denormalize the input frames
    input_sequence_denorm = input_sequence[0] * 255.0
    
    # Prepare images to display
    # Create a figure
    plt.figure(figsize=(15, 3))
    
    # Display the last input frame
    plt.subplot(1, 6, 1)
    plt.imshow(input_sequence_denorm[-1, :, :, 0], cmap='gray')
    plt.title("Last Input Frame")
    plt.axis('off')
    
    # Display predicted frames
    for i in range(5):
        plt.subplot(1, 6, i+2)
        plt.imshow(predicted_frames[i, :, :, 0], cmap='gray')
        plt.title(f"Predicted Frame {i+1}")
        plt.axis('off')
    
    # Save the figure to a buffer
    buf = io.BytesIO()
    plt.savefig(buf, format='png')
    plt.close()
    buf.seek(0)
    img = Image.open(buf)
    
    return img

# Create the Gradio interface
def update_video_list(class_name):
    """
    Update the list of videos when the class is changed.
    """
    return gr.Dropdown.update(choices=video_dict[class_name], value=video_dict[class_name][0])

with gr.Blocks() as demo:
    gr.Markdown("# Future Frame Prediction from Preprocessed Videos")
    gr.Markdown("Select a class and a video to predict future frames.")
    
    with gr.Row():
        class_input = gr.Dropdown(
            label="Select Class", 
            choices=selected_classes, 
            value=selected_classes[0]
        )
        video_input = gr.Dropdown(
            label="Select Video", 
            choices=video_dict[selected_classes[0]],  # Set initial choices
            value=video_dict[selected_classes[0]][0]  # Set initial value
        )

    # Update the video list based on selected class
    class_input.change(fn=update_video_list, inputs=class_input, outputs=video_input)
    
    predict_button = gr.Button("Predict Future Frames")
    output_image = gr.Image(type="pil", label="Predicted Frames")

    # Define the function to be called on button click
    predict_button.click(
        fn=predict_future_frames_from_npy,
        inputs=[class_input, video_input],
        outputs=output_image
    )

# Launch the interface
demo.launch()
```
